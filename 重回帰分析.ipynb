```python
import pandas as pd
import statsmodels.api as sm
import pickle

for i in range(2, 48):

    # Excelデータ読み込み
    df=pd.read_excel("分析用.xlsx", sheet_name=f"Sheet{i}")

    # Excel側に存在する列名
    excel_cols=set(df.columns)

    # 元の候補列リスト
    candidate_cols=["縦横比","縦横比sq","45°縦横比","45°縦横比sq","黒画素エントロピー","黒画素エントロピーsq","黒画素上下比","黒画素上下比sq","黒画素左右比","黒画素左右比sq","黒画素対角比","黒画素対角比sq","距離エントロピー","距離エントロピーsq","角度標準偏差","角度標準偏差sq","x距離差","x距離差sq","y距離差","y距離差sq","部分系長さ","部分系長さsq","部分系面積","部分系面積sq","ストローク長さ","ストローク長さsq","ストローク面積","ストローク面積sq"]

    # 共通する列だけ抽出
    selected_cols=[col for col in candidate_cols if col in excel_cols]

    # Xデータフレーム作成
    X=df[selected_cols]

    # 目的変数用データフレーム作成
    Y=df[["評価値"]]

    # 切片項追加
    X=sm.add_constant(X)

    # 重回帰モデル作成
    model=sm.OLS(Y,X).fit()

    # モデルをpklファイルで保存
    with open(f"model{i}.pkl","wb") as f:
        pickle.dump(model,f)

    # モデル作成結果を表示
    print(model.summary())

    # モデル作成結果をcsvファイルで出力
    # モデル作成結果を文字列に変換
    summary_txt=model.summary().as_text()
    # 作成結果の文字列を改行で分割しリスト化
    summary_lines=summary_txt.split("\n")
    # 作成結果のリストをデータフレーム化
    summary_df=pd.DataFrame(summary_lines)
    # 作成結果のcsvファイルとして出力
    summary_df.to_csv(f"summary{i}.csv",index=False,header=False,encoding="utf-8-sig")
```
